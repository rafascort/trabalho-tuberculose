# 02 - Analise Exploratoria (EDA)
Exploramos o conjunto de **treino** para entender o alvo, os dados faltantes e a
relacao das variaveis com o abandono.

```python
# bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.float_format = '{:.3f}'.format
import sys; sys.path.append("../src")   # acesso a preprocessing.py
DATA = "../data/processed"
```

```python
treino = pd.read_csv(f"{DATA}/treino.csv")
print("treino:", treino.shape)
treino[["idade_anos","NU_CONTATO","ltfu"]].describe()
```

## Distribuicao do alvo (desbalanceamento)

```python
# o abandono e a classe minoritaria -> cuidado com a acuracia
treino["ltfu"].value_counts(normalize=True).rename({0:"Cura",1:"Abandono"}).plot(
    kind="bar", color=["#4c72b0","#c44e52"])
plt.ylabel("Proporcao"); plt.title("Distribuicao do alvo (treino)"); plt.show()
```

## Dados faltantes (missings)

```python
# percentual de faltantes por coluna (top 12)
faltantes = (treino.isna().mean()*100).sort_values(ascending=False).head(12)
faltantes.plot(kind="barh", color="#dd8452"); plt.gca().invert_yaxis()
plt.xlabel("% faltante"); plt.title("Colunas com mais dados faltantes"); plt.show()
```

## Variaveis x abandono

```python
# idade por desfecho
sns.boxplot(data=treino, x="ltfu", y="idade_anos")
plt.title("Idade por desfecho (0=Cura, 1=Abandono)"); plt.show()
```

```python
# taxa de abandono por categoria (alcool, drogas, tratamento observado, HIV)
for col in ["AGRAVALCOO","AGRAVDROGA","TRAT_SUPER","HIV"]:
    print(f"\n== % de abandono por {col} ==")
    print((treino.groupby(col)["ltfu"].mean()*100).round(1))
```

**Leitura:** alcoolismo, uso de drogas, HIV+ e ausencia de tratamento diretamente
observado (TRAT_SUPER) tendem a se associar a mais abandono - coerente com a literatura.
